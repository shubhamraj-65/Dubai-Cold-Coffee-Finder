```python
from geopy.distance import geodesic
user_location=(25.240,55.315)
spot_location=(25.165,55.305)
geodesic(user_location,spot_location).km
```

```python
def calculate_distance(row):
    lat = row["lat"]
    lng = row["lng"]
    spot_location =(lat,lng)
    dis= geodesic(user_location,spot_location).km
    return round(dis,2)
df["distance_km"] =df.apply(calculate_distance,axis=1)
```

```python
from datetime import datetime
datetime.now().strftime("%H:%M")
```

```python
def is_open(row):
    current_time= datetime.now().strftime("%H:%M")
    open_time=row["opening_time"]
    close_time=row["closing_time"]
    if open_time <= current_time <= close_time:
        return "open"
    else:
        return "closed"
```
